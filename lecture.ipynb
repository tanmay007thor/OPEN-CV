```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
logo = cv2.imread('./image/logo.png', 0)  
fshift = np.fft.fft2(logo)
fshift = np.fft.fftshift(fshift)  
magnitude_spectrum = np.abs(fshift)
plt.subplot(121), plt.imshow(logo, cmap='gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(np.log(1 + magnitude_spectrum), cmap='gray')
plt.title('Log Magnitude Spectrum'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img1 = cv2.imread('./image/white_box.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('./image/white_box2.jpeg', cv2.IMREAD_GRAYSCALE)
result = cv2.matchTemplate(img1, img2, cv2.TM_CCOEFF_NORMED)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
plt.subplot(131), plt.imshow(img1, cmap='gray')
plt.title('Image 1'), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(img2, cmap='gray')
plt.title('Image 2'), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(result, cmap='gray')
plt.title('Correlation Result'), plt.xticks([]), plt.yticks([])
plt.show()
print("Maximum correlation coefficient:", max_val)
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('./image/Big_Tiger_Cub.jpg', cv2.IMREAD_COLOR)

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

equalized_img = cv2.equalizeHist(gray_img)

plt.figure(figsize=(10, 5))

plt.subplot(231), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image (Color)'), plt.xticks([]), plt.yticks([])

plt.subplot(232), plt.imshow(gray_img, cmap='gray')
plt.title('Original Image (Grayscale)'), plt.xticks([]), plt.yticks([])

plt.subplot(233), plt.imshow(equalized_img, cmap='gray')
plt.title('Equalized Image (Grayscale)'), plt.xticks([]), plt.yticks([])

# Plot images with histograms
plt.subplot(234), plt.hist(img.flatten(), bins=256, color='r', range=[0, 256], histtype='step')
plt.title('Histogram (Color)')

plt.subplot(235), plt.hist(gray_img.flatten(), bins=256, color='r', range=[0, 256], histtype='step')
plt.title('Histogram (Grayscale)')

plt.subplot(236), plt.hist(equalized_img.flatten(), bins=256, color='r', range=[0, 256], histtype='step')
plt.title('Equalized Histogram (Grayscale)')

plt.show()
```
